```python
#EXERCISE A

import numpy as np #imports the numpy as np

A = np.array([
    [2, 4, 6],
    [24, 16, 8],     # This row contains integer values arranged in a nested list
    [25, 50, 75]
])


print("Shape:", A.shape)  #in this line Output will show the dimensions of the array 3 rows, 3 columns

print("Max using method:", A.max())  #finds and returns the largest value in the entire array


print("Max using function:", np.max(A)) #finds the maximum value

print("\nArray A:") #the \n creates a new blank line before printing
print(A)  #displays the entire array in matrix format
```

```python
#EXERCISE B

import numpy as np #this will import the numpy library

B = np.ones((5, 5), dtype=int) # creates a 5x5 array with all values = 1

print("Array B:")
print(B) # Display the 5x5 array
print("Shape:", B.shape) # Returns the dimensions of the array 5 rows, 5 columns.
print("Size:", B.size)  # shows total number of values
print("Dimensions:", B.ndim) # it prints how many dimensions(which is 2)
```

```python
#EXERCISE C

import numpy as np

C = np.arange(-5, 6)  # Create an array from -5 to 5. The number 6 is the stop value, which is not included.
# np.arange() always stops before the stop value, so 5 is included but 6 is not.

result = C * 2  # multiply each value in the array by 2

print("Original array:", C)  # displays original numbers
print("Multiplied by 2:", result)  # Shows the result after multiplying by 2
```
