# Challenge 1


```python
total = 0
count = 0

while True:
    try:
        num_str = input('Enter a number: ')
        if num_str == 'done':
            break

        num = float(num_str)
        total += num
        count += 1
    except ValueError:
        print('Invalid input')

if count > 0:
    average = total / count
    print(f'{total} {count} {average}')
else:
    print('No numbers were entered.')
```

```python
numbers = [3, 1, 4, 1, 5, 9, 2, 6, 5]

max_value = max(numbers)
min_value = min(numbers)

print(f"Original list: {numbers}")
print(f"Maximum value: {max_value}")
print(f"Minimum value: {min_value}")
```

```python
numbers = [3, 1, 4, 1, 5, 9, 2, 6, 5]

if not numbers:
    print("The list is empty.")
else:
    # Initialize max and min with the first element of the list
    current_max = numbers[0]
    current_min = numbers[0]

    # Iterate through the rest of the list
    for num in numbers[1:]:
        if num > current_max:
            current_max = num
        if num < current_min:
            current_min = num

    print(f"Original list: {numbers}")
    print(f"Maximum value (manual): {current_max}")
    print(f"Minimum value (manual): {current_min}")
```

```python
import os, json

# Path to the notebook you successfully opened
filename = "/content/drive/MyDrive/IS110_BPERFORMANCE_MADRIAL/Copy of 01 Python Fundamentals, Part 1 (Workbook).ipynb"
output_file = "IS110_BPERFORMANCE_MADRIAL_extracted.py"

# Open and parse the notebook JSON
with open(filename, 'r', encoding='utf-8') as f:
    notebook_content = json.load(f)

# Extract only code cells
code_cells = []
for cell in notebook_content.get("cells", []):
    if cell.get("cell_type") == "code":
        code_cells.append("".join(cell.get("source", [])))

# Save them into a .py file
with open(output_file, "w", encoding="utf-8") as out:
    out.write("\n\n".join(code_cells))

print(f" Extraction complete! Code saved to '{output_file}'")
```
